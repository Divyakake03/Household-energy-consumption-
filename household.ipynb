```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import datatable as dt
import pandas as pd

# Load your datatable frame
data = dt.fread("C:/Users/Admin/household/household_power_consumption.csv")

# Convert to pandas DataFrame
df = data.to_pandas()
# Convert columns to numeric (float), handling errors by coercing invalid entries to NaN
cols_to_convert = ["Global_active_power", "Global_reactive_power", "Voltage",
                   "Global_intensity", "Sub_metering_1","Sub_metering_2" ,"Sub_metering_3",]

df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Handle NaN values (e.g., by filling them with 0 or other values)
df[cols_to_convert] = df[cols_to_convert].fillna(0)

# Convert from float to int after handling NaN
df[cols_to_convert] = df[cols_to_convert].astype(int)

# Verify the changes
df.info()
# Only use numeric columns, excluding the target
target_col = 'Global_active_power'
numeric_cols = df.select_dtypes(include='number').columns.tolist()
features = [col for col in numeric_cols if col != target_col]

# Create subplots
fig, ax = plt.subplots(ncols=3, nrows=2, figsize=(20, 10))
ax = ax.flatten()

index = 0
for col in features:
    sns.regplot(x=df[col], y=df[target_col], ax=ax[index])
    ax[index].set_title(f'{col} vs Global_active_power')
    index += 1
    if index >= len(ax):
        break

plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
print("Features being plotted:", features)
print("Number of plots:", len(ax))
plt.savefig("plot_output.png")
plt.show()




```
